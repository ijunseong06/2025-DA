```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))

x = np.linspace(-10, 10, 100)
y1 = x ** 2
y2 = 3 * x + 2.5

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("그래프")
plt.grid(True)
plt.show();
```
